# DBSCAN

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

## Import the Dataset

```python
df = pd.read_csv('cluster2.csv')
print("Dimensions of the data = {}".format(df.shape))

X = df.values
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.05, min_samples=20)
# eps : Maximum radius of the neighborhood, 이웃과의 최대 반지름
# min_samples : the number of points in a neighborhood to be a core point, 코어 포인트가 될 이웃의 포인트 수

y_pred = dbscan.fit_predict(X) # perform clustering & output the cluster index for each data point
print(y_pred[:10])
# -1 : Outliers
```

```python
cmap = "tab10"

for i, (eps, min_samples) in enumerate(((0.05, 20), (0.06, 20), (0.06, 15), (0.06, 6))):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    y_pred = dbscan.fit_predict(X)
    plt.figure(i+1, figsize=(5, 5))
    plt.scatter(X[:, 0], X[:, 1], cmap=cmap, s=4, c=y_pred)
    plt.title("eps: {}, min_samples: {}".format(eps, min_samples))
    
plt.show()
```
