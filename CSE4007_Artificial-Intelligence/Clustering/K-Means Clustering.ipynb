# K-Means Clustering

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

## Import the Dataset

```python
df = pd.read_csv('cluster2.csv')
print("Dimensions of the data = {}".format(df.shape))
```

```python
df[:5]
```

```python
X = df.values
X[:5]
```

## Plotting the Dataset

```python
# Set the size of the figure
plt.figure(figsize=(5, 5)) # width, height

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], s = 4) # x, y coordinates | size of each points

plt.show()
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters = 4, random_state = 0)

cmap = "tab10"
```

### Prediected Cluster Indices

```python
y_pred = k_means.fit_predict(X)
print(y_pred[:10])
```

### Cluster Centers

```python
print(k_means.cluster_centers_)
```

### Plotting Clustering Results

```python
plt.figure(figsize = (5, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=4, cmap=cmap)

plt.show()

# K-means 클러스터링은 다른 사이즈, Non-spherical 의 특징을 가지기 힘들기 때문에 결과가 다음과 같다.
```
